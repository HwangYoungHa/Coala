```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('data/train.csv')
```

```python
# hue : 구분 기준, kind : 데이터 수집 방식(max, min, mean, count 등)
sns.catplot(data=train, x='Survived', hue='Sex', kind='count')
```

```python
# Pclass에 따른 생존자 인원 출력
sns.catplot(data=train, x='Survived', hue='Pclass', kind='count')
```

```python
# 선착장별 생존자와 사망자
sns.catplot(data=train, x='Survived', hue='Embarked', kind='count')
```

```python
# 생존자, 사망자의 나이 분포
sns.catplot(data=train, x='Age', hue='Survived', kind='count')
```
