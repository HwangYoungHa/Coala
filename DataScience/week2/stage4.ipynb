```python
plt.plot(['Math', 'Korean', 'English'], [90, 100, 50])
plt.plot(['Math', 'Korean', 'English'], [70, 60, 90]) # 서로 다름 -> 알아서 구분

plt.xlabel('Subjects') # x축 label 붙이기
plt.ylabel('Score') # y축 label 붙이기
plt.title('Midterm') # title

plt.legend(['Choi', 'Kim']) # line 구분
plt.show()
```

```python
x = ['Math', 'Korean', 'English']
y = [90, 100, 50]

plt.xlabel('Subjects')
plt.ylabel('Score')
plt.bar(x, y, width=0.3, color='#970618') # 그래프 타입(막대 그래프), 두께 및 색깔 변경
plt.show()
```

```python
# 막대차트
x = ['math', 'korean', 'english']
y = [90, 100, 50]
plt.xlabel('subjects') # x값의 라벨
plt.ylabel('score') # y값의 라벨

plt.bar(x, y, width=0.3, color = 'black') # 막대그래프 출력
plt.show()
```

```python
plt.plot([1, 2, 3], [90, 100, 50], marker='+', markersize=10, linestyle='--')
# marker : 점의 모양 / linestyle : 선의 모양
plt.grid(True)
plt.axis([1, 3, 80, 100]) # 1 ~ 3 : x범위 / 80 ~ 100 : y범위
```

```python
import pandas as pd
```

```python
train = pd.read_csv('data/train.csv')

train
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set()
```

```python
 # train.csv의 내용물 지정 및 x, y축 지정
sns.relplot(data=train, x='Age', y='Survived', hue='Sex', aspect=4)
```

```python
 # train.csv의 내용물 지정 및 x, y축 지정
sns.relplot(data=train, x='Age', y='Survived', hue='Sex', aspect=4)
#aspect : 가로길이 = 세로길이 * 4
```

```python
# 카테고리 (분포 출력)
sns.catplot(data=train, x='Pclass', y='Fare', aspect=4, jitter=False)
# jitter : 기본값 설정(True / False)
```

```python
sns.pairplot(data=train, hue='Sex', # hue : 구분 기준
            x_vars=['Pclass', 'Age', 'Fare'], # 행과 열 설정
            y_vars=['Survived', 'Age', 'Fare'], height=3)
```

```python
sns.catplot(data=train, x='Pclass', y='Fare', hue='Sex', kind='box')
# 박스 형식으로 출력
# 점 : 실제로 유효 x값
# 막대 : 유효값(최대 / 최솟값)
```
