```python
import pandas as pd
```

```python
df = pd.read_csv('data/iris.csv')
df.head(10)
```

```python
# 데이터 프레임의 정보 출력
df.info()
```

```python
df.describe() # 해당 특징들의 데이터 정보 출력
```

```python
# 그래프를 위한 라이브러리
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set()
```

```python
sns.catplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', aspect=2)
```

```python
# 꽃잎의 길이를 기준으로 그래프 그리기
sns.catplot(data=df, x='PetalLengthCm', y='PetalWidthCm', hue='Species', aspect=2)
```

```python
# kdeplot를 통한 분포도 그리기
facet = sns.FacetGrid(df, hue='Species', aspect=2) # aspect : 그리드의 길이(가로)
facet.map(sns.kdeplot, 'SepalLengthCm')
facet.add_legend()
plt.show()
```

```python
# catplot
sns.catplot(kind='box', data=df, x='Species', y='SepalLengthCm', aspect=2)
```

```python
# box plot으로 종별 꽃받침 길이 범위 파악
sns.catplot(kind='box', data=df, x = 'Species', y='SepalLengthCm', aspect=2)
```
