```python
import codecs

positive = []
negative= []
posneg = []

# ê¸ì •ì–´, ë¶€ì •ì–´ ì‚¬ì „ ë¶ˆëŸ¬ì˜¤ê¸°
pos = open("./positive-words.txt", 'r', encoding='UTF-8')
neg = open("./negative-words.txt", 'r', encoding='UTF-8')
```

```python
# ê¸ì •ì–´, ë¶€ì •ì–´ë¥¼ listì— ì¶”ê°€
while True:
    line = pos.readline()
    line = line.replace('\n', '')
    positive.append(line)
    posneg.append(line)
    if not line: break
pos.close()

while True:
    line = neg.readline()
    line = line.replace('\n', '')
    negative.append(line)
    posneg.append(line)
    if not line: break
neg.close()
```

```python
# ê° ë¬¸ìž¥ì„ í† ëŒ€ë¡œ ê¸ì •, ë¶€ì • íŒë³„
pos_plabel = [] # í•´ë‹¹ ë¬¸ìž¥ì˜ ê¸ì • ë¶€ì • íŒë³„
pos_nlabel = []
neg_plabel = []
neg_nlabel = []
```

```python
i = 0
for review in df['Positive_Review']:
    pos_plabel.append(0)
    pos_nlabel.append(0)
    # ì‚¬ì „ì„ í† ëŒ€ë¡œ ì¡°ì‚¬
    for pword in positive: 
        if pword in review:
            pos_plabel[i] += 1
    for nword in negative:
        if nword in review:
            pos_nlabel[i] += 1
    i += 1
```

```python
# pos_plabel # ê¸ì •ì ì¸ ë‹¨ì–´ì˜ ë¹ˆë„ìˆ˜
```

```python
# pos_nlabel # ë¶€ì •ì ì¸ ë‹¨ì–´ì˜ ë¹ˆë„ìˆ˜
```

```python
i = 0
for review in df['Negative_Review']:
    neg_plabel.append(0)
    neg_nlabel.append(0)
    # ì‚¬ì „ì„ í† ëŒ€ë¡œ ì¡°ì‚¬
    for pword in positive: 
        if pword in review:
            neg_plabel[i] += 1
    for nword in negative:
        if nword in review:
            neg_nlabel[i] += 1
    i += 1
```

```python
df['pos_plabel']=pos_plabel
df['pos_nlabel']=pos_nlabel
df['neg_plabel']=neg_plabel
df['neg_nlabel']=neg_nlabel

df
```

# plabelê³¼ nlabelì„ ê·¸ëž˜í”„ë¡œ ê·¸ë¦¬ê¸°

```python
df['p_ratio']=df['pos_plabel']/df['pos_nlabel']
df['n_ratio']=df['neg_plabel']/df['neg_nlabel']

df
```
