```python
import codecs

positive = []
negative= []
posneg = []

# 긍정어, 부정어 사전 불러오기
pos = open("./positive-words.txt", 'r', encoding='UTF-8')
neg = open("./negative-words.txt", 'r', encoding='UTF-8')
```

```python
# 긍정어, 부정어를 list에 추가
while True:
    line = pos.readline()
    line = line.replace('\n', '')
    positive.append(line)
    posneg.append(line)
    if not line: break
pos.close()

while True:
    line = neg.readline()
    line = line.replace('\n', '')
    negative.append(line)
    posneg.append(line)
    if not line: break
neg.close()
```

```python
# 각 문장을 토대로 긍정, 부정 판별
pos_plabel = [] # 해당 문장의 긍정 부정 판별
pos_nlabel = []
neg_plabel = []
neg_nlabel = []
```

```python
i = 0
for review in df['Positive_Review']:
    pos_plabel.append(0)
    pos_nlabel.append(0)
    # 사전을 토대로 조사
    for pword in positive: 
        if pword in review:
            pos_plabel[i] += 1
    for nword in negative:
        if nword in review:
            pos_nlabel[i] += 1
    i += 1
```

```python
# pos_plabel # 긍정적인 단어의 빈도수
```

```python
# pos_nlabel # 부정적인 단어의 빈도수
```

```python
i = 0
for review in df['Negative_Review']:
    neg_plabel.append(0)
    neg_nlabel.append(0)
    # 사전을 토대로 조사
    for pword in positive: 
        if pword in review:
            neg_plabel[i] += 1
    for nword in negative:
        if nword in review:
            neg_nlabel[i] += 1
    i += 1
```

```python
df['pos_plabel']=pos_plabel
df['pos_nlabel']=pos_nlabel
df['neg_plabel']=neg_plabel
df['neg_nlabel']=neg_nlabel

df
```

# plabel과 nlabel을 그래프로 그리기

```python
df['p_ratio']=df['pos_plabel']/df['pos_nlabel']
df['n_ratio']=df['neg_plabel']/df['neg_nlabel']

df
```
